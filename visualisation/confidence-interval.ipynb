```python
import numpy as np
from scipy.stats import bootstrap

error_res = np.array((-10, 0, 0, 9, 25, 
                      38, 59, 56, 73, 73,
                      102, 109))

error_off = np.array((-83, -52, -29, -25, -19, -15, -7, -5, -0.7, 12, 14, 32, 41, 46, 64, 72, 75,
                      113, 160))
```

```python
from random import choices

statistics = []
for i in range(1000):
    sample = choices(error_res, k=30)
    stat = np.mean(sample)
    statistics.append(stat)

statistics = np.sort(statistics)
```

```python
mean = np.mean(error_res)
std = np.std(error_res)
print(mean - 2*std, mean+2*std)
```
